<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(2*-np.pi, 2*np.pi, 200)

tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
#code here
fig, ax = plt.subplots(1,3, figsize=(18,8))
[axi.set_axis_off() for axi in ax.ravel()]
labels = ["tanX", "sinX", "cosX"]
l1 = ax[0].plot(tan, color="blue")
l2 = ax[1].plot(sin, color="red")
l3 = ax[2].plot(cos, color="green")
fig.legend([l1, l2, l3], labels=labels,
           loc="upper center")
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
#code here
fig, ax = plt.subplots(figsize=(8,7))
ax.plot(x, y)
ins2 = ax.inset_axes([-0.2, -0.2, 0.3, 0.3])
ins2.plot(x, y2, color="red")
ins3 = ax.inset_axes([-0.2, 0.9, 0.3, 0.3])
ins3.plot(x, y3, color="#FFDF50")
ins4 = ax.inset_axes([0.9, 0.9, 0.3, 0.3])
ins4.plot(x, y4, color="black")
ins5 = ax.inset_axes([0.9, -0.2, 0.3, 0.3])
ins5.plot(x, y5, color="green")
```
