<h2>Soal 4: Perbandingan Kuantitatif Barplot - Grouping Visualisasi Dengan Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menghasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
import numpy as np

# set width of bar
barWidth = 0.05
fig = plt.subplots(figsize =(12, 8))
data = []
# set height of bar
for row in range(0, len(df_group_sales)):
    data.append(df_group_sales.values.tolist()[row])

# Set position of bar on X axis
br1 = np.arange(len(data[0]))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]
br4 = [x + barWidth for x in br3]
br5 = [x + barWidth for x in br4]
br6 = [x + barWidth for x in br5]
br7 = [x + barWidth for x in br6]
br8 = [x + barWidth for x in br7]
br9 = [x + barWidth for x in br8]
br10 = [x + barWidth for x in br9]
br11 = [x + barWidth for x in br10]
br12 = [x + barWidth for x in br11]
 
# Make the plot
plt.bar(br1, data[0], color ='r', width = barWidth,
        edgecolor ='grey', label ='Action')
plt.bar(br2, data[1], color ='g', width = barWidth,
        edgecolor ='grey', label ='Adventure')
plt.bar(br3, data[2], color ='b', width = barWidth,
        edgecolor ='grey', label ='Fighting')
plt.bar(br4, data[3], color ='c', width = barWidth,
        edgecolor ='grey', label ='Misc')
plt.bar(br5, data[4], color ='m', width = barWidth,
        edgecolor ='grey', label ='Platform')
plt.bar(br6, data[5], color ='y', width = barWidth,
        edgecolor ='grey', label ='Puzzle')
plt.bar(br7, data[6], color ='k', width = barWidth,
        edgecolor ='grey', label ='Racing')
plt.bar(br8, data[7], color ='#FF00AA', width = barWidth,
        edgecolor ='grey', label ='Role-Playing')
plt.bar(br9, data[8], color ='#00DDAA', width = barWidth,
        edgecolor ='grey', label ='Shooter')
plt.bar(br10, data[9], color ='#1000AA', width = barWidth,
        edgecolor ='grey', label ='Simulation')
plt.bar(br11, data[10], color ='#FFDDFF', width = barWidth,
        edgecolor ='grey', label ='Sports')
plt.bar(br12, data[11], color ='#FFFFAB', width = barWidth,
        edgecolor ='grey', label ='Strategy')
 
# Adding Xticks
plt.xlabel('Region Sales', fontsize = 8)
plt.ylabel('Mean Sales', fontsize = 8)
plt.xticks([r + 0.3 for r in range(len(data[0]))],
        ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales'])
 
plt.legend()
plt.show()
```

Dari chart batang diatas bisa kita ketahui bahwa rata-rata penjualan video game terbanyak ada di region NA. Dan juga untuk genre terpopuler pada tiap region adalah:

* NA = Platform
* EU = Shooter
* JP = Role Playing
* Others = Shooter
