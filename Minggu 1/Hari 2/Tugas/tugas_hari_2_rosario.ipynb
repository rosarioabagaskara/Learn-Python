# Data Structure

<h2>Soal 1: Perbedaan Data Structure</h2>

Jawab Pertanyaan di bawah ini:

Jelaskan perbedaan dari List, Tuple, Set dan Dictionary!

List: List merupakan tipe data yang bisa digunakan untuk menampung beberapa data dan beberapa tipe data. List merupakan tipe data mutable dimana kita bisa menambah, menghapus, atau mengupdate value yang ada didalam list.

Tuple: Tuple sama seperti List, merupakan tipe data yang digunakan untuk menampung beberapa data dan beberapa tipe data. Bedanya, pada Tuple kita tidak bisa menambah, menghapus, atau mengupdate data yang ada didalam Tuple.

Set: Set merupakan suatu tipe data yang menampung beberapa data tetapi data yang disimpan adalah unique. Sehingga ketika terdapat 2 data yang bernilai sama dimasukkan kedalam set, data yang masuk hanya 1 data saja. Data dalam set bisa diubah dan memiliki sifat seperti himpunan.

Dictionary: Seperti List dan Tuple, kita dapat menggunakan Dictionary untuk menampung beberapa data dan tipe data. Namun, pada dictionary kita perlu memetakan Key dan Values dari data yang dimasukkan.

## Soal 6: List Comprehension

Gunakan metode **list comprehension** untuk mencari anggota dari S yang habis di bagi 2, kemudian assign hasilnya dalam bentuk list ke dalam variabel T.

```python
S = [0, 1, 4, 9, 16, 25, 36, 49, 64, 81]

T = [i for i in S if i % 2 == 0]

print(T)
```

Expected Output:

[0, 4, 16, 36, 64]

<h2>Soal 10: Nested Dictionary</h2>
    
Lengkapi kode untuk menghasilkan suatu output yang di harapkan

```python
from itertools import count


country = { 
           'spain': { 'capital':'madrid', 'population':46.77 },
           'france': { 'capital':'paris', 'population':66.03 },
           'germany': { 'capital':'berlin', 'population':80.62 },
           'norway': { 'capital':'oslo', 'population':5.084 } 
         }

# berapa populasi dari kota german?
print(country["germany"]["population"])


# update data baru, yaitu negara indonesia dengan capital jakarta dan poulasi 250
country["indonesia"] = {
    "capital" : "jakarta",
    "population" : 250
}


print(country)
```

Expected Output:

80.62

{'spain': {'capital': 'madrid', 'population': 46.77}, 'france': {'capital': 'paris', 'population': 66.03}, 'germany': {'capital': 'berlin', 'population': 80.62}, 'norway': {'capital': 'oslo', 'population': 5.084}, 'indonesia': {'capital': 'jakarta', 'population': 250}}

<h2>Soal 11: Loop Dictionary</h2>
    
Lengkapi kode untuk menghasilkan suatu output yang di harapkan

```python
country = { 
           'spain': { 'capital':'madrid', 'population':46.77 },
           'france': { 'capital':'paris', 'population':66.03 },
           'germany': { 'capital':'berlin', 'population':80.62 },
           'norway': { 'capital':'oslo', 'population':5.084 },
           'indonesia' : {'capital':'jakarta', 'population':250}
         }

for key, value in country.items():
    print('Ibukota '+str(key)+' adalah ' + str(value["capital"]))
```

Expected Output:
    
Ibukota spain adalah madrid

Ibukota france adalah paris

Ibukota germany adalah berlin

Ibukota norway adalah oslo

Ibukota indonesia adalah jakarta

## Soal 12: Remove Duplicate using set

## Soal 13: Mengubah dan menghapus anggota set

Ubahlah dan hapus anggota set sehingga mendapatkan hasil yang diinginkan


```python
set_1 = {1, 2, 3, 4, 5}

#tambahkan anggota set dengan nilai {6,7,8}
set_1.update([6,7,8])

#hapus nilai anggota set 8
set_1.remove(8)

print(set_1)
```
