# Pivoting

```python
data = {
    'foo' : ['one']*3 + ['two']*3,
    'bar' : ['A', 'B', 'C', 'A', 'B', 'C'],
    'baz' : [1, 2, 3, 4, 5, 6],
    'zoo' : ['x', 'y', 'z', 'q', 'w', 't']
}
df = pandas.DataFrame(data)
print(df)

# using pivot
pivot_df = pandas.pivot(df, index='foo', columns='bar', values='baz')

print(pivot_df)

#using pivot_table
pivot_table_df= pandas.pivot_table(df, index='foo', columns='bar', values='baz', aggfunc='sum')

print(pivot_table_df)
```
