<h2>Soal 3: DataFrame vs Series</h2>

- Apa perbedaan Series dengan list dan dictionary
- Apa perbedaan dataframe dan series
- Buatlah suatu data series

Jawab: 
* Perbedaan antara Series dan list adalah, Series merupakan objek 1 dimensi dimana data didalamnya bisa menerima berbagai tipe data. Namun, tipe data yang ada didalam series, harus seragam. Sedangkan tipe data didalam list tidak harus seragam. 

    Perbedaan series dan dictionary adalah series dapat memiliki key yang duplikat dan series dapat melakukan slicing pada data yang terdapat didalam series. DIctionary tidak bisa memiliki key yang duplikat dan tidak dapat dilakukan slicing.

* Perbedaan dataframe dan series adalah, Dataframe merupakan objek 2 dimensi, sedangkan series merupakan objek 1 dimensi. Data yang ditampilkan oleh dataframe berbentuk baris dan kolom, sedangkan series menampilkan baris. 

```python
# code here
series_ = pd.Series(['Motor', 'Mobil', 'Sepeda'])
print(series_)
```
