# Time Series

```python
# DateTime Python

import datetime

date = datetime.date(year=2022, month=10, day=28)
time = datetime.time(hour=12, minute=25, second=49,microsecond=1)

dt = datetime.datetime(year=2022, month=10, day=28, hour=12, minute=25, second=49,microsecond=1)

print(f'date is {date}')
print(f'time is {time}')

print(f'datetime is {dt}')
```

```python
#time delta

td = datetime.timedelta(weeks=2, days=5, hours=10,

minutes=20, seconds=6.73,

milliseconds=99, microseconds=8)

print(f'time delta is {td}')

# Jumlahkan timedelta ke date & datetime
print('new date is', date + td)
print('new datetime is', dt + td)
```
